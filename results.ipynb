```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
def cosinesimilarity(matrix1, matrix2):
    # Flatten the matrices
    matrix1 = sorted(matrix1)
    matrix2 = sorted(matrix2)
    matrix1_flat = np.array(matrix1).flatten().reshape(1, -1)
    matrix2_flat = np.array(matrix2).flatten().reshape(1, -1)
    return cosine_similarity(matrix1_flat, matrix2_flat)[0][0]
```
